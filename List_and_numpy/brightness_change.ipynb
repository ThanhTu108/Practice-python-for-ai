```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread('nature.png', 1)
img = img[:, :, ::-1]
print(img.dtype)
img = img - 50  #broadcasting 
plt.imshow(img)
```

# Clip and Where functions in NumPy


## Clip() function

```python
img2 = cv2.imread('nature.png')
img2 = img2.astype(float)
img2 = np.clip(img2 + 50, a_min=0, a_max=255)
img2 = img2.astype(np.uint8)
img2 = img2[:, :, ::-1]
plt.imshow(img2)
```

## where() function

```python
img3 = cv2.imread('nature.png')

img3 = img3.astype(float)
img3 = img3 - 100
img3 = np.where(img3 < 0, 0, img3)
img3 = img3.astype(np.uint8)
img3 = img3[:, :, ::-1]
plt.imshow(img3)
```
