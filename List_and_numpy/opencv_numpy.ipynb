### import libraries


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('nature.png', 1)   #color in cv2 is BGR, plt: RGB
img_rgb_1 = img[:, :, [2,1,0]]
img_rgb_2 = img[:, :, ::-1]
img_rgb_3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb_1)
print(type(img_rgb_1))
print(img_rgb_1.shape)
```

```python
plt.imshow(img_rgb_2),
print(type(img_rgb_2))
print(img_rgb_2.shape)
```

```python
plt.imshow(img_rgb_3),
print(type(img_rgb_3))
print(img_rgb_3.shape)
```

```python
img = cv2.imread('nature_grayscale_1.jpg', 0)
plt.imshow(img, cmap='gray')
print(type(img))
print(img.shape)
```
