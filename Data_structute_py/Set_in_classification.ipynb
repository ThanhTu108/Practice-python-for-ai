# Convert text -> Number

```python
a_file = open("Data_set_in_classification.txt", "r")
data = a_file.read()
a_file.close()
print(data)
```

## Preprocessing


```python
def  standardlize(data):
    # Func aim to convert string -> list with many elements from data
    data = data.replace(".", "")
    data = data.replace(",", "")
    data = data.replace("-", " ")
    data = data.lower()
    data = data.split()
    # # convert list to set to ensure all elements are unique
    # data = set(data)
    return data
```

```python
def vectorization(data):
    # convert list to set to ensure all elements are unique
    data = set(data)
    # convert set to dictionary to have key, value pairs
    return {value:index for index, value in enumerate(data)}
```

```python

data = standardlize(data)
print(data)
print()

data = vectorization(data)
print(data)
```

## text to number

```python
word = "python"
text_to_number = data[word]
print(text_to_number)
```
