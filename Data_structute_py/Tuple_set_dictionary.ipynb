# 1D list solution

```python
student_list = ["Vinh", "n", "Ton", "Lm", "Tm", "Chuyn"]
row_class = 3
col_class = 2
def fine_student(studen_list, r, c):
    
    index = (r-1)*col_class + (c - 1)
    return studen_list[index]

fine_student(student_list, 3, 1)
```

EX: Data protection in process

```python
def quadratic_equation(a, b, c):
    ''' This func aims to at solving the quadratic equation 
    with a, b, c are three parameter, a !=0
    '''

    delta = b*b - 4*a*c

    if delta < 0:
        return ()
    elif delta == 0:
        x = (-b / (2*a))
        return (x,)
    else:
        x1 = (-b + delta**0.5/(2*a))
        x2 = (-b - delta**0.5/(2*a))
        return (x1, x2)
```

```python
result = quadratic_equation(1,2,3)

print(f"result = {result}, {type(quadratic_equation(1,2,3))}")

result = quadratic_equation(1,2,0)

print(f"result = {result}, {type(quadratic_equation(1,2,3))}")

result = quadratic_equation(4,4,1)

print(f"result = {result}, {type(quadratic_equation(1,2,3))}")
```

# Dictionary Motivation

## Structure
dictionary_name = {key1:valua1, ..... , keyn:valuen}

## Create a dictionary

```python
a_dict = {i:i for i in range(10)}
print(a_dict)
```

```python
# zip (tuple)

tuple1 = (1,2,3)    #key
tuple2 = (4,5,6)    #value

a_dict = dict(zip(tuple1, tuple2))
print(type(a_dict))
print(a_dict)
```

```python
# zip (set)

set1 = {1,2,3}    #key
set2 = {4,5,6}    #value

a_dict = dict(zip(set1, set2))
print(type(a_dict))
print(a_dict)
```

```python
# zip (list)

list1 = [1,2,3]    #key
list2 = [4,5,6]    #value

a_dict = dict(zip(list1, list2))
print(type(a_dict))
print(a_dict)
```

## Copy a dictionary

```python
parameter = {"learning_rate": 0.1,
             "metric": "Accuracy"}

a_copy = parameter.copy()
a_copy["learning_rate"] = 0.4
print(parameter)
print(a_copy)
print(type(a_copy))
```

```python
import copy
#shallow copy -> con tr 
d1 = {'a': [1,2], 'b': [3,4]}
d2 = d1.copy()

d2['a'][0] = 3
print(d1)
print(d2)

d1 = {'a': [1,2], 'b': [3,4]}
d2 = copy.deepcopy(d1)

d2['a'][0] = 3
print(d1)
print(d2)
```
